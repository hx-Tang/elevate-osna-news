```python
import pandas as pd
df=pd.read_csv('D:\\news\\training_data\\factchecks.csv')
df1=pd.read_csv('D:\\news\\training_data\\youtube.csv')[['social_id','comment_tokens']]
df=df[df.site=='youtube']
df2=pd.merge(df,df1,on=['social_id'],how='inner')
df2=df2[['social_id','comment_tokens','ruling','ruling_val']]
df2.columns=['id','text','cat','label']
df2['text'] = df2['text'].fillna('')
df2.head()
```

```python
list=df2['text'].values.tolist()
from sklearn.feature_extraction.text import TfidfVectorizer
import re
tv=TfidfVectorizer(analyzer='word',token_pattern=r'[^0-9_\W]+',min_df=1,stop_words = 'english')
tv.fit_transform(list)
word=tv.get_feature_names()
model=tv.fit(list)
weight=model.transform(list)
print("→Length of all terms:",len(word))
print(word)
print("→Length of all weights:",weight.shape[1])
print(weight)
```

```python
weight.todense()
```

```python
import numpy as np
word_list=np.array(word)
col_sum = weight.sum(axis=0).A1
# Sort sums in descending order, and return the top 10 indices.
top_indices = np.argsort(col_sum)[::-1][:200]
# print('top_indices=', top_indices)
term = word_list[top_indices]
freq = col_sum[top_indices]
[x for x in zip(term, freq)]
```

```python
from sklearn.feature_extraction import DictVectorizer
from collections import Counter
dv = DictVectorizer()
feature_dicts=[]
words_to_track=['trump','obama','vote','president','fake','funny']
for i, row in df2.iterrows():
        features = {}
        token_counts = Counter(re.sub('[0-9_\W]+', '', row['text']).split())
        for w in words_to_track:
            features[w] = token_counts[w]
        feature_dicts.append(features)
X = dv.fit_transform(feature_dicts)
print(X.shape)
```

```python
dv.vocabulary_
```

```python
for word, idx in dv.vocabulary_.items():
    print('%20s\t%d' % (word, X[:,idx].sum()))
```

```python
class_names = set(df2.label)
y=np.array(df2.label)
for word, idx in dv.vocabulary_.items():
    for class_name in class_names:
        class_idx = np.where(y==class_name)[0]
        print('%20s\t%20s\t%d' % (word, class_name, X[class_idx, idx].sum()))
```

```python
from sklearn.linear_model import LogisticRegression
lrmodel = LogisticRegression(solver='lbfgs', multi_class='auto',max_iter=5000)
lrmodel.fit(X, y)
```

```python
lrmodel.classes_
```

```python
lrmodel.coef_
```

```python
coef = lrmodel.coef_
features=dv.get_feature_names()
for ci, class_name in enumerate(lrmodel.classes_):
    print('Coefficients for %s' % class_name)
    display(pd.DataFrame([coef[ci]], columns=features))
    print('→Top features for class %s :' % class_name)
    for fi in coef[ci].argsort()[::-1]: # descending order.
        print('%20s\t%.2f' % (features[fi], coef[ci][fi]))
    print("\n")
```

```python
def accuracy(truth, predicted):
    return len(np.where(truth==predicted)[0]) / len(truth)

predicted = lrmodel.predict(X)
print('accuracy on training data=%.3f' % accuracy(y, predicted))
```

```python
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score

kf = KFold(n_splits=5, shuffle=True, random_state=42)
accuracies = []
for train, test in kf.split(X):
    lrmodel.fit(X[train], y[train])
    pred = lrmodel.predict(X[test])
    accuracies.append(accuracy_score(y[test], pred))
    
    
print('Accuracy over all cross-validation folds:\n %s' % str(accuracies))
print('mean=%.2f std=%.2f' % (np.mean(accuracies), np.std(accuracies)))
```
